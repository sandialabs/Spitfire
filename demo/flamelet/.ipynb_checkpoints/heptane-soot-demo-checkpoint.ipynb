```python
from spitfire.chemistry.mechanism import ChemicalMechanismSpec
from spitfire.chemistry.flamelet import Flamelet
from time import perf_counter
import numpy as np
```

The following cell will build an adiabatic flamelet library. From here we will run transient extinction calculations of nonadiabatic flamelets. We run only one in this demonstration.

```python
m = ChemicalMechanismSpec(cantera_xml='mechanisms/nHeptane.LiuHewsonChenPitsch04_highT.xml', group_name='gas')

pressure = 101325.

air = m.stream(stp_air=True)
air.TP = 300, pressure

c7 = m.stream('TPY', (485., pressure, 'NXC7H16:1'))

flamelet_specs = {'mech_spec': m,
                  'pressure': pressure,
                  'oxy_stream': air,
                  'fuel_stream': c7,
                  'grid_points': 128,
                  'grid_cluster_intensity': 4.,
                  'include_enthalpy_flux': True}

flamelet_for_grid = Flamelet(max_dissipation_rate=0, initial_condition='equilibrium', **flamelet_specs)

x_range = np.hstack([0., np.logspace(-4, np.log10(160.), 20)])

z_range = flamelet_for_grid.mixfrac_grid
x_grid, z_grid = np.meshgrid(x_range, z_range, indexing='ij')

slfm_table = dict()
species_to_save = ['C2H2', 'H', 'OH', 'O2']
keys = ['temperature', 'density', 'viscosity', 'heat release rate'] + ['mass fraction ' + x for x in species_to_save]
for k in keys:
    slfm_table[k] = np.ndarray((x_range.size, z_range.size))

state_list = []

running_state = None
for idx, x in enumerate(x_range):
    flamelet_specs['initial_condition'] = running_state if running_state is not None else 'equilibrium'
    flamelet_specs['max_dissipation_rate'] = x
    flamelet = Flamelet(**flamelet_specs)
    cput0 = perf_counter()
    flamelet.compute_steady_state()
    dcput = perf_counter() - cput0
    running_state = flamelet.final_interior_state
    state_list.append(np.copy(flamelet.final_interior_state))
    print('converged flamelet {:4}/{:4} (chi_max = {:8.1e} 1/s) in {:5} iterations ({:7.2f} s), T_max = {:6.1f}'
          .format(idx + 1, x_range.size, x, flamelet.iteration_count, dcput, np.max(running_state)))

    for k in keys:
        if k == 'viscosity':
            flamelet.insitu_process_cantera_method('viscosity')
        else:
            flamelet.insitu_process_quantity(k)
    data_dict = flamelet.process_quantities_on_state(flamelet.final_state)
    for k in keys:
        slfm_table[k][idx, :] = data_dict[k]
```

The next cell writes a function that computes soot source terms and adds them to the table.

This means they would be written to HDF5 just like any other quantity!

Also note how simple NumPy makes this and how there is no boilerplate associated with writing data to file.

```python
def compute_soot_source_terms(table):
    from cantera import gas_constant as R
    
    rho = table['density']
    T = table['temperature']
    c_C2H2 = rho * table['mass fraction C2H2'] / m.gas.molecular_weights[m.gas.species_index('C2H2')]
    c_O2 = rho * table['mass fraction O2'] / m.gas.molecular_weights[m.gas.species_index('O2')]
    c_OH = rho * table['mass fraction OH'] / m.gas.molecular_weights[m.gas.species_index('OH')]
    
    soot_density = 1800.
    N_avagadro = 6.02e26
    vol_fac = (6. / np.pi / soot_density / N_avagadro) ** (1. / 3.)
    area_fac = vol_fac * vol_fac * np.pi * N_avagadro

    coag_coeff = np.sqrt(vol_fac * 24. * R / soot_density / N_avagadro)
    surf_coeff = 11.7e3 * area_fac
    surf_O2_coeff = 500. * area_fac
    neoh_fac = 0.04
    surf_OH_coeff = 4.2325 / neoh_fac * area_fac

    table['nucleation rate'] = 54. * np.exp(-21100. / T) * c_C2H2
    table['coagulation rate'] = coag_coeff * np.sqrt(T) * rho ** 2
    table['surface growth rate'] = surf_coeff * np.exp(-12000. / T) * c_C2H2 * rho
    table['surface O2 rate'] = surf_O2_coeff * np.sqrt(T) * np.exp(-20000. / T) * c_O2 * rho
    table['surface OH rate'] = surf_OH_coeff * neoh_fac * np.sqrt(T) * c_OH * rho
```

```python
compute_soot_source_terms(slfm_table)
```

```python
idx_for_h = 18
x_for_h = x_range[idx_for_h]
state_for_h = state_list[idx_for_h]

flamelet_specs['initial_condition'] = state_for_h
flamelet_specs['max_dissipation_rate'] = x
flamelet_specs['heat_transfer'] = 'nonadiabatic'
flamelet_specs['use_scaled_heat_loss'] = True
flamelet_specs['convection_coefficient'] = 1.e7
flamelet_specs['radiative_emissivity'] = 0.

flamelet = Flamelet(**flamelet_specs)

for k in keys:
    if k == 'viscosity':
        flamelet.insitu_process_cantera_method('viscosity')
    else:
        flamelet.insitu_process_quantity(k)
flamelet.insitu_process_quantity('enthalpy')

flamelet.integrate_for_heat_loss(write_log=True,
                                 log_rate=100,
                                 max_time_step=1.e-1,
                                 transient_tolerance=1.e-12,
                                 first_time_step=1.e-9)
```

```python
from scipy.interpolate import interp1d

t_vec = flamelet.solution_times
nt = t_vec.size

h_data = flamelet.trajectory_data('enthalpy')

h_st_vec = np.zeros_like(flamelet.solution_times)
for i in range(nt):
    h_st_vec[i] = interp1d(flamelet.mixfrac_grid, h_data[i, :])(zst)
```

```python
dh_st_total = h_st_vec[-1] - h_st_vec[0]
dh_st_target = -10.e3

interval = 1 #int(np.floor(h_st_vec.size / int(np.ceil(dh_st_total / dh_st_target))))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t_vec * 1e6, (h_st_vec - h_st_vec[0]) / 1.e6, '-', label='all times')
ax.plot(t_vec[::interval] * 1e6, (h_st_vec[::interval] - h_st_vec[0]) / 1.e6, 'o', markerfacecolor='none', label='for table')
ax.set_xlabel('t (us)')
ax.set_ylabel('$\\gamma_{st}$ (MJ/kg)')
ax.grid()
ax.legend(loc='best')
plt.show()

print('max dh_st = {:.1f} kJ/kg'.format(np.max(h_st_vec[::interval][1:] - h_st_vec[::interval][:-1]) / 1.e3))
```

```python
indices = np.array(list(range(h_st_vec.size)))[::interval]
h_st_vec_table = h_st_vec[::interval]
nh = h_st_vec_table.size
heatloss_table = dict()
for k in keys + ['enthalpy']:
    heatloss_table[k] = np.ndarray((nh, z_range.size))
    
for idx in range(nh):
    for k in keys + ['enthalpy']:
        heatloss_table[k][idx, :] = flamelet.trajectory_data(k)[indices[idx], :]
        
dh_st_grid, z_hst_grid = np.meshgrid((h_st_vec_table - h_st_vec_table[0]) / 1.e6, z_range, indexing='ij')
```

```python
from scipy.interpolate import interp2d
from numpy.matlib import repmat

key = 'temperature'

quantity = heatloss_table[key]

zst = m.stoich_mixture_fraction(c7, air)
print(zst)
z = flamelet.mixfrac_grid
nz = z.size
enthalpy = flamelet.trajectory_data('enthalpy')
gamma_st_range = np.zeros(enthalpy.shape[0])
h_st_ad = interp1d(z, enthalpy[0, :])(zst)
for i in range(gamma_st_range.size):
    gamma_st_range[i] = (interp1d(z, enthalpy[i, :])(zst) - h_st_ad) / 1.e6
zg_key_interp = interp2d(z_range, gamma_st_range, quantity)

full_gamma_data = (enthalpy - repmat(enthalpy[0, :], enthalpy.shape[0], 1)).ravel() / 1.e6
full_z_data = repmat(flamelet.mixfrac_grid, enthalpy.shape[0], 1).ravel()
full_quantity = flamelet.trajectory_data(key).ravel()

def gamma_inverter(gamma, z, zst):
    if z == 0 or z == 1:
        return 0.
    elif z < zst:
        return zst / z * gamma
    elif z > zst:
        return (1 - zst) / (1 - z) * gamma
    else:
        return gamma
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(full_z_data, full_gamma_data, c=full_quantity, marker='.', cmap=plt.get_cmap('magma'))
gi = 200
gamma_gi = full_gamma_data[gi*nz:(gi+1)*nz]
ax.plot(z, gamma_gi, 'c-', linewidth=2, label='actual')
gi2 = 100
gamma_gi2 = full_gamma_data[gi2*nz:(gi2+1)*nz]
ax.plot(z, gamma_gi2, 'b-', linewidth=2, label='bleh')

z_test = 0.45
gamma_test = interp1d(z, gamma_gi)(z_test)
gamma_test_st = gamma_inverter(gamma_test, z_test, zst)
gamma_test_profile = np.zeros_like(z)
gamma_test_profile[z <= zst] = z[z <= zst] / zst * gamma_test_st
gamma_test_profile[z > zst] = (1-z[z > zst])/(1-zst) * gamma_test_st
ax.plot(z, gamma_test_profile, 'w--', linewidth=2, label='presumed')
ax.plot(z_test, gamma_test, 'ws-', markersize=10, markeredgecolor='k', label='test point')

ax.set_xlabel('Z')
ax.set_ylabel('$\\gamma$')
fig.colorbar(mappable=ax.collections[0])
ax.set_title(key)
ax.grid(True)
ax.legend(loc='best')
ax.set_xlim([0, 1])
```

```python
struc_reconstructed_key = np.zeros_like(full_quantity)
for i in range(full_gamma_data.size):
    gamma_st = gamma_inverter(full_gamma_data[i], full_z_data[i], zst)
    struc_reconstructed_key[i] = zg_key_interp(full_z_data[i], gamma_st)
struc_error = np.abs(struc_reconstructed_key - full_quantity)
print(np.max(struc_error))

fig, ax = plt.subplots(1, 1)
ax.scatter(full_z_data, full_gamma_data, c=struc_reconstructed_key, marker='.', cmap=plt.get_cmap('magma'))
ax.set_xlabel('Z')
ax.set_ylabel('$\\gamma$')
fig.colorbar(mappable=ax.collections[0])
ax.set_title('reconstructed ' + key)
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(full_z_data, full_gamma_data, c=struc_error, marker='.', cmap=plt.get_cmap('magma'))
ax.set_xlabel('Z')
ax.set_ylabel('$\\gamma$')
fig.colorbar(mappable=ax.collections[0])
ax.set_title('reconstruction error in ' + key)
```

```python
from pyearth import Earth as Mars
```

```python
model = Mars(max_degree=3, max_terms=1000, check_every=1, minspan=1, endspan=10, verbose=1)
x_mars = np.vstack([full_z_data, full_gamma_data]).T
model.fit(x_mars, full_quantity)
mars_reconstructed_key = model.predict(x_mars)
mars_error = np.abs(mars_reconstructed_key - full_quantity)
print(np.max(mars_error))
```
